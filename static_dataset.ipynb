```python
import pandas as pd
import os
import re
```

# Dados de piramide de idade de 2019

```python
if os.path.exists("/Users/thrabbit/Documents/corona_virus/input/pop_pyramid.csv"):
    df_pyramid = pd.read_csv("/Users/thrabbit/Documents/corona_virus/input/pop_pyramid.csv")
else:
    df_pyramid = pd.DataFrame(columns=['Country', 'Age 0-9', 'Age 10-19', 'Age 20-29', 'Age 30-39', 'Age 40-49', 'Age 50-59', 'Age 60-69', 'Age 70-79', 'Age >80'])
```

```python
directory = "/Users/thrabbit/Documents/corona_virus/input/pop_pyramid/"
pattern = "^.*(?=(-2019\.csv))"

for file in os.listdir(directory):
    if file.endswith(".csv"):
        df = pd.read_csv(directory + file)
        df = df.set_index('Age')
        df['Population'] = df['M'] + df['F']
        df = df.transpose()

        country = re.search(pattern, file).group(0)
        
        country_df = pd.DataFrame(data={'Country':country, 
               'Age 0-9':[df.loc['Population']['0-4'] + df.loc['Population']['5-9']], 
               'Age 10-19':[df.loc['Population']['10-14'] + df.loc['Population']['15-19']], 
               'Age 20-29':[df.loc['Population']['20-24'] + df.loc['Population']['25-29']],
               'Age 30-39':[df.loc['Population']['30-34'] + df.loc['Population']['35-39']],
               'Age 40-49':[df.loc['Population']['40-44'] + df.loc['Population']['45-49']], 
               'Age 50-59':[df.loc['Population']['50-54'] + df.loc['Population']['55-59']], 
               'Age 60-69':[df.loc['Population']['60-64'] + df.loc['Population']['65-69']], 
               'Age 70-79':[df.loc['Population']['70-74'] + df.loc['Population']['75-79']], 
               'Age >80':[df.loc['Population']['80-84'] + df.loc['Population']['90-94'] + df.loc['Population']['95-99'] + df.loc['Population']['100+']]})
        
        df_pyramid = df_pyramid.append(country_df, ignore_index=True)
```

# Calculando a população total

```python
df_pyramid['# People'] = df_pyramid.iloc[:, 1:10].sum(axis=1)/10e3
```
